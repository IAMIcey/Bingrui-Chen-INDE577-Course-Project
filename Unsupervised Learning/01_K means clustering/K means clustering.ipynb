# K means clustering


K-means clustering is a widely used unsupervised learning algorithm for data clustering. It partitions the dataset into K distinct, non-overlapping subgroups (or clusters) based on the features provided, where each data point belongs to the cluster with the nearest mean. This algorithm is particularly effective in situations where you need to identify intrinsic groupings among your data.

## The K-means Algorithm

The algorithm involves the following steps:

**Initialization:** K initial centroids are chosen at random from the dataset.

**Assignment Step:** Each data point is assigned to its nearest centroid, based on the Euclidean distance, or another distance metric.

**Update Step:** The centroids of the clusters are recalculated as the mean of all data points assigned to that cluster.

**Iterative Convergence:** The assignment and update steps are repeated iteratively until the centroids no longer move significantly or a maximum number of iterations is reached.

# Application

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set theme 
sns.set_theme()
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.head(10)
```

```python
from sklearn.cluster import KMeans
X = df.loc[:, ["median_income", "latitude", "longitude"]]
X.head()
```

```python
kmeans = KMeans(n_clusters=6)
X["Cluster"] = kmeans.fit_predict(X)
X["Cluster"] = X["Cluster"].astype("category")

X.head()
```

```python
sns.relplot(
    x="longitude", y="latitude", hue="Cluster", data=X, height=6,
);
```

## Advantages

Simplicity and Efficiency: K-means is easy to understand and implement, and it is computationally efficient for large datasets.

Adaptability: Works well with a variety of datasets and has been successfully applied in various domains.

## Challenges

Choosing K: The number of clusters (K) must be specified beforehand, and finding the optimal K is not straightforward.

Sensitivity to Initialization: The algorithm is sensitive to the initial choice of centroids. Different initializations may lead to different clusters.

Assumption of Spherical Clusters: K-means assumes that clusters are spherical and equally sized, which might not always be the case.

Outliers: The presence of outliers can significantly distort the shape of clusters.
