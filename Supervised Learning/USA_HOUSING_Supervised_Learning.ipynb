```python
X=df[['Avg. Area Income', 'Area Population']]
y=df['Price']
```

```python
X = df[['Avg. Area Income', 'Area Population']].values
y = df['Price'].values

# Convert targets to binary for the purpose of demonstration
y= np.where(y > y.mean(), 1, -1)
```

```python
class Perceptron(object):
    def __init__(self, eta = .5, epochs=50):
        self.eta = eta
        self.epochs = epochs
        
    def train(self, X, y):
        self.w_ = np.random.rand(1 + X.shape[1])
        
        self.errors_ = []
        
        for _ in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (self.predict(xi) - target)
                self.w_[:-1] -= update*xi
                self.w_[-1] -= update
                errors += int(update != 0)
            if errors == 0:
                return self
            else:
                self.errors_.append(errors)
            
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w_[:-1]) + self.w_[-1]
    
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
# Instantiate one instance of the Perceptron class
clf = Perceptron(epochs = 1_000)

# Call the train method to train the weights and bias of the given instance
clf.train(X, y)
```

```python
# Predict labels on the features in X
y_hat = clf.predict(X)

# Compare y_hat and y
print(y == y_hat)
```

```python
plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = clf)
plt.title("My First Perceptron", fontsize = 18)
plt.xlabel("Avg. Area Income", fontsize = 15)
plt.ylabel("Area Population", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plt.plot(range(1, len(clf.errors_)+1), clf.errors_, 
         marker = "o",
         label = "error plot",)
plt.xlabel("Iterations")
plt.ylabel("Missclassifications")
plt.legend()
plt.show()
```
